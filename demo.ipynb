```
df = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/georgemichaeldagogomaynard@gmail.com/data.csv")
df.createOrReplaceTempView('df_query')
query1= ''' SELECT 
price,
num_reviews,
Rating
FROM df_query
WHERE Rating >0.5 '''

query2 = ''' SELECT
price, 
num_lectures, 
published_timestamp,
num_subscribers,
subject
FROM df_query 
WHERE published_timestamp LIKE '2014%'
'''

new_df = spark.sql(query1)
newer_df = spark.sql(query2)

newer_df.show()
```
